```python
from scipy.special import erf
def loc_F_fric_viscosity_model(T, y, yF, yWt=0.03, yC=0.4):
  ''' Calculates the viscosity as a function of dissolved
  water and crystal content (assumes crystal phase is incompressible)/
  Does not take into account fragmented vs. not fragmented (avoiding
  double-dipping the effect of fragmentation).
  '''
  # Calculate pure melt viscosity (Hess & Dingwell 1996)
  yWd = yWt - y
  yM = 1.0 - y
  yL = yM - (yWd + yC)
  mfWd = yWd / yL # mass concentration of dissolved water
  mfWd = np.where(mfWd <= 0.0, 1e-8, mfWd)
  log_mfWd = np.log(mfWd*100)
  log10_vis = -3.545 + 0.833 * log_mfWd
  log10_vis += (9601 - 2368 * log_mfWd) / (T - 195.7 - 32.25 * log_mfWd)
  # Prevent overflowing float
  log10_vis = np.where(log10_vis > 300, 300, log10_vis)
  meltVisc = 10**log10_vis
  # Calculate relative viscosity due to crystals (Costa 2005).
  alpha = 0.999916
  phi_cr = 0.673
  gamma = 3.98937
  delta = 16.9386
  B = 2.5
  # Compute volume fraction of crystal at equal phasic densities
  # Using crystal volume per (melt + crystal + dissolved water) volume
  phi_ratio = np.clip((yC / yM) / phi_cr, 0.0, None)
  erf_term = erf(
    np.sqrt(np.pi) / (2 * alpha) * phi_ratio * (1 + phi_ratio**gamma))
  crysVisc = (1 + phi_ratio**delta) * ((1 - alpha * erf_term)**(-B * phi_cr))
  
  viscosity = meltVisc * crysVisc
  return viscosity

locale = r"C:\Users\Fredric\Documents\Volcano\quail_volcano\src\compressible_conduit_steady"
import os
os.chdir(locale)
import steady_state

f = steady_state.SteadyState(np.linspace(0,10000,100), 1e5, 100e6, input_type="p",
                             override_properties={
                               "tau_f": 1e0, # 1e-2,
                               "tau_d": 1e0, # 1e-2,
                              #  "yWt": 0.05,
                               "yC": 0.4,
                               "rho0_magma": 2.6e3,
                             }, skip_rootfinding=True)

params = {
  "T0": 950+273.15,
  "conduit_radius": 10,
  "tau_d": 10.0,
  "tau_f": 1.0,
  "vf_g_crit": 0.8,
  "solubility_n": f.solubility_n,
  "solubility_k": f.solubility_k,
  "K": 10e9,
  "rho_m0": 2.7e3,
  "p_m0": 5e6,
  "R_wv": f.mixture.waterEx.R,
  "F_fric_viscosity_model": loc_F_fric_viscosity_model,
}
```

```python
# Alias interpolants
r_interpolants = q5_interp_raw
# Create choking event
p0 = 40e6
u0 = 2.9
M_eps = 1e-2
def event_isothermal_choked(x, q2_scaled) -> np.array:
  ''' Returns isothermal mach squared '''
  p = q2_scaled[0,...]
  u = q2_scaled[1,...]
  yWv = r_interpolants(x)[1,...]
  rho = split_solver.rho_p(yWv, p)
  ptilde = split_solver.p_ptilde(rho, yWv)[1]
  return split_solver.isothermal_mach_squared(u, yWv, p, ptilde) - (1.0 - M_eps) * (1.0 - M_eps)
event_isothermal_choked.terminal = True
# event_isothermal_choked.direction = 1.0
def time_indep_jac_scalar(x, q2, r_interpolants):
# Interpolate solution for time-dependent quantities
  r = r_interpolants(x)
  # Unpack quantities
  p         = q2[0:1,...]
  u         = q2[1:2,...]
  rho_prime = r[0:1,...]
  yWv       = r[1:2,...]
  yWt       = r[2:3,...]
  yC        = r[3:4,...]
  yF        = r[4:5,...]
  # Return scalarized RHS_jac
  return split_solver.RHS_jac(p, u, yWv, yWt, yC, yF)[0,...]

# Solve IVP
solve_with_method_no_jac = lambda method_name: scipy.integrate.solve_ivp(
  lambda x, q: split_solver.time_indep_RHS(
    x, q, r_interpolants, is_debug_mode=False),
  (0, 1000),
  np.array([p0, float(u0)]),
  method=method_name,
  dense_output=True,
  max_step=np.inf,
  rtol=1e-6,
  atol=1e-8,
  events=[event_isothermal_choked,])
solve_with_method = lambda method_name: scipy.integrate.solve_ivp(
  lambda x, q: split_solver.time_indep_RHS(
    x, q, r_interpolants, is_debug_mode=False),
  (0, 1000),
  np.array([p0, float(u0)]),
  method=method_name,
  jac=lambda x, q: time_indep_jac_scalar(x, q, r_interpolants),
  dense_output=True,
  max_step=np.inf,
  rtol=1e-6,
  atol=1e-8,
  events=[event_isothermal_choked,])
```

```python
''' Set up numerical Jacobian of RHS vector function. '''
yWv = split_solver.q5[1,...]
yF = split_solver.q5[4,...]
p = split_solver.q2[0,...]

q5_interp_raw = scipy.interpolate.PchipInterpolator(split_solver.x,
                                        split_solver.q5,
                                        extrapolate=True,
                                        axis=1)

def q5_interp_scalar(xq):
  ''' Scalar returns'''
  return np.where(xq <= x.max(),
                  q5_interp_raw(xq),
                  q5_interp_raw(x.max()))

def compute_numjac(x:float, q2_scalar:np.array, eps=1e-5):
  scale1 = 1e5 # pressure scale
  scale2 = 1 # velocity scale
  _e1 = np.array([scale1, 0])
  _e2 = np.array([0, scale2])
  fp1 = split_solver.time_indep_RHS(x, (q2_scalar + 0.5 * eps * _e1), q5_interp_scalar)
  fm1 = split_solver.time_indep_RHS(x, (q2_scalar - 0.5 * eps * _e1), q5_interp_scalar)
  fp2 = split_solver.time_indep_RHS(x, (q2_scalar + 0.5 * eps * _e2), q5_interp_scalar)
  fm2 = split_solver.time_indep_RHS(x, (q2_scalar - 0.5 * eps * _e2), q5_interp_scalar)
  _numjac = np.array([(fp1 - fm1)/(scale1*eps), (fp2 - fm2) / (scale2*eps)]).T
  # _numjac, (fp2 - fm2)/_eps, fp1
  return _numjac
```

```python
# Exact Jacobian of RHS (Ainv@b) (Prototype, OK)

def dvdp(self, yWv, p):
  rhoM = self.rhoM_p(p)
  return -(yWv * self.R_wv * self.T0 / (p * p)
           + (1 - yWv) / (rhoM * rhoM) * self.rho_m0 / self.K)

def fn_Jpp(split_solver, yWv, yWt, yC, p):
  self = split_solver

  rho = self.rho_p(yWv, p)
  _, ptilde = self.p_ptilde(rho, yWv)
  # Compute dp/dyWv group (const rho, T)
  Pi = (-p * (self.K * rho / self.rho_m0) 
              + (p + self.K - self.p_m0) * rho * self.R_wv * self.T0
            ) / (2*p - ptilde)
  _drhodp = - rho * rho * dvdp(self, yWv, p)
  _g1 = -2 / (2*p - ptilde)* Pi
  _g2 = (self.K / self.rho_m0 * (1.0 - yWv) + self.R_wv * self.T0 * yWv) / (2*p - ptilde) * _drhodp * Pi
  _g3 = 1/rho * _drhodp * Pi
  _g4 = rho / (2*p - ptilde) * (-self.K / self.rho_m0 + self.R_wv * self.T0)
  ddp_dpdYWv = _g1 + _g2 + _g3 + _g4

  yL = 1.0 - (yWt + yC)
  p_like = np.clip(p, 1e-16, None)
  yWdHat = np.clip(self.solubility_k * p_like ** self.solubility_n * yL, 0, yWt)
  _h1 = ddp_dpdYWv * (yWt - yWdHat - yWv) / self.tau_d
  _h2 = -self.dp_dyw(rho, p, ptilde) / self.tau_d * (
    self.solubility_n * self.solubility_k * p_like**(self.solubility_n - 1) * yL)
  _h2 = np.where(yWt >= self.solubility_k * p ** self.solubility_n * yL, _h2, 0.0)
  return _h1 + _h2

def exact_fjac(split_solver):
  ''' Jacobian of RHS (without mass matrix inversion) '''
  p = split_solver.q2[0,...]
  u = split_solver.q2[1,...]
  yWv = split_solver.q5[1,...]
  yWt = split_solver.q5[2,...]
  yC = split_solver.q5[3,...]
  yF = split_solver.q5[4,...]

  mu = split_solver.F_fric_viscosity_model(
    split_solver.T0, yWv, yF, yWt=yWt, yC=yC)
  F_fric = 8.0 * mu / (split_solver.conduit_radius * split_solver.conduit_radius) \
              * u * np.clip(1.0 - yF/(1.0 - yWv), 0.0, 1.0)
  rho = split_solver.rho_p(yWv, p)
  Juu = -8.0 * mu / rho / (split_solver.conduit_radius * split_solver.conduit_radius) \
              * np.clip(1.0 - yF/(1.0 - yWv), 0.0, 1.0)
  Jup = -8.0 * mu / (split_solver.conduit_radius * split_solver.conduit_radius) \
              * u * np.clip(1.0 - yF/(1.0 - yWv), 0.0, 1.0) \
              * dvdp(split_solver, yWv, p)
  Jpu = np.zeros_like(yF)
  Jpp = fn_Jpp(split_solver, yWv, yWt, yC, p)
  J = np.zeros((p.size, 2, 2))
  J[:,0,0] = Jpp
  J[:,0,1] = Jpu
  J[:,1,0] = Jup
  J[:,1,1] = Juu
  return J

_exactjacs = exact_fjac(split_solver)
_exactjacs.shape

# Compute Jacobian of Ainv*f using product rule:
#   d_k(Ainv_ij) @ f_j + Ainv_ij @ df_jk
rho = split_solver.rho_p(yWv, p)
_, ptilde = split_solver.p_ptilde(rho, yWv)
cT2 = split_solver.isothermal_sound_speed_squared(yWv, p, ptilde)
Ainv = np.array([[u, -rho*cT2], [-1/rho, u]]) / (u*u - cT2)

# rank2-rank2 part
_prod22 = np.einsum("ijz, zjk -> zik", Ainv, _exactjacs)

def ddp_cT2(split_solver, yWv, yWt, yC, p):
  self = split_solver
  rho = self.rho_p(yWv, p)
  _, ptilde = self.p_ptilde(rho, yWv)
  yM = 1.0 - yWv
  # Compute drho/dp (const yWv)
  _drhodp = - rho * rho * dvdp(self, yWv, p)
  # Compute d/dp (2p - pTilde)
  _ddp_group = 2 - (self.K / self.rho_m0 * yM + self.R_wv * self.T0 * yWv) * _drhodp
  _num = (2*p - ptilde) * (yM * self.K / self.rho_m0 + yWv * self.R_wv * self.T0) \
    - (yM * p * self.K / self.rho_m0 + yWv * (p + self.K - self.p_m0) * self.R_wv * self.T0) \
    * _ddp_group
  return _num / ((2*p - ptilde) * (2*p - ptilde))

mu = split_solver.F_fric_viscosity_model(
  split_solver.T0, yWv, yF, yWt=yWt, yC=yC)
F_fric = 8.0 * mu / (split_solver.conduit_radius * split_solver.conduit_radius) \
            * u * np.clip(1.0 - yF/(1.0 - yWv), 0.0, 1.0)
rho = split_solver.rho_p(yWv, p)
p_like = np.clip(p, 1e-16, None)
yL = 1.0 - (yWt + yC)
yWdHat = np.clip(split_solver.solubility_k * p_like ** split_solver.solubility_n * yL, 0, yWt)
dp_dyw = split_solver.dp_dyw(rho, p, ptilde)
f = np.stack([
  dp_dyw * (yWt - yWdHat - yWv) / split_solver.tau_d,
  -F_fric/rho - 9.8,
], axis=1)
_drhodp = - rho * rho * dvdp(split_solver, yWv, p)

# dAinv/dp @ f == Ainv @ dA/dp @ Ainv @ f as row
# dAinv/du @ f = Ainv @ dA/du @ Ainv @ f as row
_dAdp_01 = _drhodp * split_solver.isothermal_sound_speed_squared(yWv, p, ptilde) + rho * ddp_cT2(split_solver, yWv, yWt, yC, p)
_dAdp_10 = dvdp(split_solver, yWv, p)
# Compute Ainv @ f
_Ainvf = np.einsum("ij..., ...j -> ...i", Ainv, f)
# Compute dA/dp @ Ainv @ f explicitly (sparse off diagonal matrix)
_dAdp_Ainvf = np.stack([
  _dAdp_01 * _Ainvf[...,1],
  _dAdp_10 * _Ainvf[...,0],
], axis=1)
# rank3-rank1 part
_prod31 = np.stack([
  -np.einsum("ij..., ...j -> ...i", Ainv, _dAdp_Ainvf),
  -np.einsum("ij..., ...j -> ...i", Ainv, _Ainvf),
], axis=2)
# # rank3-rank1 part
_JJ = _prod22 + _prod31
# _JJ.shape
_numjacs[0,...], _JJ[0,...]
```

```python
# Analytic Jacobian Packaging (Copied to .py!)

def RHS_jac(split_solver, p, u, yWv, yWt, yC):
  ''' Jacobian of RHS as Ainv@f, w.r.t state vector [p, u].
  This function proceeds by computing:
    1. (Ainv)_ij @ (d_k f_j)
  and then
    2. (d_k (Ainv)_ij) @ f_j. '''
  self = split_solver

  ''' Compute simple dependents '''
  rho = self.rho_p(yWv, p)
  _, ptilde = self.p_ptilde(rho, yWv)
  yL = 1.0 - (yWt + yC)
  p_like = np.clip(p, 1e-16, None)
  yWdHat = np.clip(self.solubility_k * p_like ** self.solubility_n * yL, 0, yWt)
  cT2 = self.isothermal_sound_speed_squared(yWv, p, ptilde)
  mu = self.F_fric_viscosity_model(
      self.T0, yWv, yF, yWt=yWt, yC=yC)
  F_fric = 8.0 * mu / (self.conduit_radius * self.conduit_radius) \
                * u * np.clip(1.0 - yF/(1.0 - yWv), 0.0, 1.0)
  # Shared derivatives
  _dvdp = self.dvdp(yWv, p)
  _dp_dyw = self.dp_dyw(rho, p, ptilde)
  _drhodp = - rho * rho * _dvdp

  ''' Compute Ainv, f '''
  #   d_k(Ainv_ij) @ f_j + Ainv_ij @ df_jk
  Ainv = np.array([[u, -rho*cT2], [-1/rho, u]]) / (u*u - cT2)
  f = np.stack([
    _dp_dyw * (yWt - yWdHat - yWv) / self.tau_d,
    -F_fric/rho - 9.8,
  ], axis=1)
  
  ''' Compute Ainv times jacobian of f (without inverse mass matrix) '''
  Juu = -8.0 * mu / rho / (self.conduit_radius * self.conduit_radius) \
              * np.clip(1.0 - yF/(1.0 - yWv), 0.0, 1.0)
  Jup = -8.0 * mu / (self.conduit_radius * self.conduit_radius) \
              * u * np.clip(1.0 - yF/(1.0 - yWv), 0.0, 1.0) \
              * _dvdp
  Jpu = np.zeros_like(yF)
  # Compute dp/dyWv group (const rho, T) with product rule terms
  _Pi = (-p * (self.K * rho / self.rho_m0) 
              + (p + self.K - self.p_m0) * rho * self.R_wv * self.T0
            ) / (2*p - ptilde)
  _g1 = -2 / (2*p - ptilde)* _Pi
  _g2 = (self.K / self.rho_m0 * (1.0 - yWv) + self.R_wv * self.T0 * yWv) / (2*p - ptilde) * _drhodp * _Pi
  _g3 = 1/rho * _drhodp * _Pi
  _g4 = rho / (2*p - ptilde) * (-self.K / self.rho_m0 + self.R_wv * self.T0)
  ddp_dpdYWv = _g1 + _g2 + _g3 + _g4
  # Compute Jpp using product rule
  _h1 = ddp_dpdYWv * (yWt - yWdHat - yWv) / self.tau_d
  _h2 = -_dp_dyw / self.tau_d * (
    self.solubility_n * self.solubility_k * p_like**(self.solubility_n - 1) * yL)
  _h2 = np.where(yWt >= self.solubility_k * p ** self.solubility_n * yL, _h2, 0.0)
  Jpp = _h1 + _h2
  # Assemble jacobian of f
  _fjac = np.zeros((p.size, 2, 2))
  _fjac[:,0,0] = Jpp
  _fjac[:,0,1] = Jpu
  _fjac[:,1,0] = Jup
  _fjac[:,1,1] = Juu
  # rank2-rank2 part: Ainv @ df
  _prod22 = np.einsum("ij..., ...jk -> ...ik", Ainv, _fjac)

  ''' Compute jacobian of A times f '''
  # dAinv/dp @ f == Ainv @ dA/dp @ Ainv @ f as row
  # dAinv/du @ f = Ainv @ dA/du @ Ainv @ f as row
  _dAdp_01 = _drhodp * cT2 + rho * self.ddp_cT2(yWv, yWt, yC, p)
  _dAdp_10 = _dvdp
  # Compute Ainv @ f
  _Ainvf = np.einsum("ij..., ...j -> ...i", Ainv, f)
  # Compute dA/dp @ Ainv @ f explicitly (sparse off diagonal matrix)
  _dAdp_Ainvf = np.stack([
    _dAdp_01 * _Ainvf[...,1],
    _dAdp_10 * _Ainvf[...,0],
  ], axis=1)
  # rank3-rank1 part: dAinv @ f
  _prod31 = np.stack([
    -np.einsum("ij..., ...j -> ...i", Ainv, _dAdp_Ainvf),
    -np.einsum("ij..., ...j -> ...i", Ainv, _Ainvf),
  ], axis=2)
  
  return _prod22 + _prod31
```

```python
def dKdp_fn(yWv, p):
  _eps = 1e-7
  dK = split_solver.K_mix(yWv, p + 0.5*_eps) \
       - split_solver.K_mix(yWv, p - 0.5*_eps)
  return dK/_eps
```

```python
def S_norm(dKdp, dudx, dpdx, rhoc):
  # Matrix 2-norm of sensitivity matrix
  a00 = dKdp * dudx
  a01 = dpdx
  a10 = -dpdx/(rhoc*rhoc)
  a11 = dudx
  A = np.stack((np.stack((a00, a01), axis=0),
                np.stack((a10, a11), axis=0),), axis=0)
  return np.linalg.norm(A)
```
